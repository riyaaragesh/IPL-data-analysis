```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("matches.csv")
```

```python
print(df.head())

print(df.info())
```

```python
print(df.isnull().sum())
```

```python
wins = df['winner'].value_counts()

print(wins)
```

```python
plt.figure(figsize=(12,6))

wins.plot(kind='bar')

plt.title("IPL Teams With Most Wins")

plt.xlabel("Teams")

plt.ylabel("Number of Wins")

plt.show()
```

```python
toss = df['toss_winner'].value_counts()

plt.figure(figsize=(12,6))

toss.plot(kind='bar')

plt.title("Teams Winning Toss Most Times")

plt.xlabel("Teams")

plt.ylabel("Toss Wins")

plt.show()
```

```python
decision = df['toss_decision'].value_counts()

print(decision)

decision.plot(kind='pie', autopct='%1.1f%%')

plt.title("Toss Decisions")

plt.ylabel('')

plt.show()
```

```python
venues = df['venue'].value_counts().head(10)

plt.figure(figsize=(12,6))

venues.plot(kind='bar')

plt.title("Top IPL Venues")

plt.xlabel("Venue")

plt.ylabel("Matches Hosted")

plt.show()
```

```python
season_matches = df['season'].value_counts().sort_index()

plt.figure(figsize=(12,6))

season_matches.plot()

plt.title("IPL Matches Per Season")

plt.xlabel("Season")

plt.ylabel("Matches")

plt.show()
```

```python
top_players = df['player_of_match'].value_counts().head(10)

plt.figure(figsize=(12,6))

top_players.plot(kind='bar')

plt.title("Top Player of Match Winners")

plt.xlabel("Players")

plt.ylabel("Awards")

plt.show()
```

```python
print("Most successful team:")
print(df['winner'].mode()[0])

print("\nMost common toss decision:")
print(df['toss_decision'].mode()[0])

print("\nTop player of the match winner:")
print(df['player_of_match'].mode()[0])
```
